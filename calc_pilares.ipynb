```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
data = {'TEMPO_FUND': [13,9],
        'SV_12M': [1000, 50000],
        'FL_VCDMERM30_R01_BC': [0, 1],
        'VL_LISEDVTOTMER_01M_BC': [9800, 100000],
        'VL_EDVTOTMER_01M_BC': [20000, 999999],
        'VL_RUREDVTOTMER_01M_BC': [150, 698000],
        'VL_FNMEDVTOTMER_01M_BC': [9856, -999],
        'VL_BNDEDVTOTMER_01M_BC': [1587, np.nan],
        'VL_IMBEDVTOTMER_01M_BC': [0, 1500],
        'VL_VEIEDVTOTMER_01M_BC': [500, 754235],
        'VL_FINIMPEDVTOTMER_01M_BC': [-999, 5000],
        'VL_VCDTOTMER_01M_BCG': [0, 652],
        'VL_PJZTOTMER_01_BCG': [0, -999]      
        }

df = pd.DataFrame(data)

df
```

```python
#Cria Variavel IND FUND
df['IND_FUND'] = 0
df.loc[df['TEMPO_FUND'] >= 12, 'IND_FUND'] = 1
```

```python
#Para Cálculo da variavel VL_ALAVANC_01M_BC, vamos tratar as variaveis que fazem parte de sua construção:
#Para isso, se o valor for negativo ou nulo, atribui-se Zero

df.loc[(df['VL_EDVTOTMER_01M_BC'] < 0) | (df['VL_EDVTOTMER_01M_BC'].isnull()), 'VL_EDVTOTMER_01M_BC'] = 0
df.loc[(df['VL_RUREDVTOTMER_01M_BC'] < 0) | (df['VL_RUREDVTOTMER_01M_BC'].isnull()), 'VL_RUREDVTOTMER_01M_BC'] = 0
df.loc[(df['VL_FNMEDVTOTMER_01M_BC'] < 0) | (df['VL_FNMEDVTOTMER_01M_BC'].isnull()), 'VL_FNMEDVTOTMER_01M_BC'] = 0
df.loc[(df['VL_BNDEDVTOTMER_01M_BC'] < 0) | (df['VL_BNDEDVTOTMER_01M_BC'].isnull()), 'VL_BNDEDVTOTMER_01M_BC'] = 0
df.loc[(df['VL_IMBEDVTOTMER_01M_BC'] < 0) | (df['VL_IMBEDVTOTMER_01M_BC'].isnull()), 'VL_IMBEDVTOTMER_01M_BC'] = 0
df.loc[(df['VL_VEIEDVTOTMER_01M_BC'] < 0) | (df['VL_VEIEDVTOTMER_01M_BC'].isnull()), 'VL_VEIEDVTOTMER_01M_BC'] = 0
df.loc[(df['VL_FINIMPEDVTOTMER_01M_BC'] < 0) | (df['VL_FINIMPEDVTOTMER_01M_BC'].isnull()), 'VL_FINIMPEDVTOTMER_01M_BC'] = 0

df['VL_ALAVANC_01M_BC'] = df['VL_EDVTOTMER_01M_BC'] - \
                          df['VL_RUREDVTOTMER_01M_BC'] - \
                          df['VL_FNMEDVTOTMER_01M_BC'] - \
                          df['VL_BNDEDVTOTMER_01M_BC'] - \
                          df['VL_IMBEDVTOTMER_01M_BC'] - \
                          df['VL_VEIEDVTOTMER_01M_BC'] - \
                          df['VL_FINIMPEDVTOTMER_01M_BC']
```

```python
df['SV_ALAVANC_MIC'] = df['SV_12M']
df['SV_ALAVANC_PQN'] = df['SV_12M']
df['SV_ALAVANC_MED'] = df['SV_12M']
df['SV_ALAVANC_GDE'] = df['SV_12M']

#Atribui Piso de Sinal Vital
df.loc[(df['SV_12M'] < 3000) | (df['SV_12M'].isnull()), 'SV_ALAVANC_MIC'] = 3000
df.loc[(df['SV_12M'] < 100000) | (df['SV_12M'].isnull()), 'SV_ALAVANC_PQN'] = 100000
df.loc[(df['SV_12M'] < 500000) | (df['SV_12M'].isnull()), 'SV_ALAVANC_MED'] = 500000
df.loc[(df['SV_12M'] < 1000000) | (df['SV_12M'].isnull()), 'SV_ALAVANC_GDE'] = 1000000
```

```python
df['VL_ALAV_SV_01M_BC_MIC'] = df['VL_ALAVANC_01M_BC']/df['SV_ALAVANC_MIC']
df['VL_ALAV_SV_01M_BC_PQN'] = df['VL_ALAVANC_01M_BC']/df['SV_ALAVANC_PQN']
df['VL_ALAV_SV_01M_BC_MED'] = df['VL_ALAVANC_01M_BC']/df['SV_ALAVANC_MED']
df['VL_ALAV_SV_01M_BC_GDE'] = df['VL_ALAVANC_01M_BC']/df['SV_ALAVANC_GDE']
```

```python
def pilar_ala_ind_mic(df):
  if df['VL_ALAVANC_01M_BC'] < 150 and df['IND_FUND'] == 1:
    return '2'
  elif df['VL_ALAVANC_01M_BC'] < 150 and df['IND_FUND'] == 0:
    return '3'
  elif df['VL_ALAV_SV_01M_BC_MIC'] < 0.01 and df['IND_FUND'] == 1:
   return '1'
  elif df['VL_ALAV_SV_01M_BC_MIC'] < 0.03 and df['IND_FUND'] == 1: 
    return '2'
  elif df['VL_ALAV_SV_01M_BC_MIC'] < 0.10 and df['IND_FUND'] == 1:
    return '3'
  elif df['IND_FUND'] == 1:
    return '4'
  elif df['VL_ALAV_SV_01M_BC_MIC'] < 0.01 and df['IND_FUND'] == 0:
    return '2' 
  elif df['VL_ALAV_SV_01M_BC_MIC'] < 0.05 and df['IND_FUND'] == 0:
    return '4'
  elif df['IND_FUND'] == 0:
    return '5'

def pilar_ala_ind_pqn(df):
  if df['VL_ALAVANC_01M_BC'] < 250 and df['IND_FUND'] == 1:
    return '1'
  elif df['VL_ALAVANC_01M_BC'] < 250 and df['IND_FUND'] == 0:
    return '1'
  elif df['VL_ALAV_SV_01M_BC_PQN'] < 0.01:
    return '1'
  elif df['VL_ALAV_SV_01M_BC_PQN'] < 0.02: 
    return '2'
  elif df['VL_ALAV_SV_01M_BC_PQN'] < 0.50:
    return '3'
  else:
    return '4'

def pilar_ala_ind_med(df):
  if df['VL_ALAVANC_01M_BC'] < 250 and df['IND_FUND'] == 1:
    return '1'
  elif df['VL_ALAVANC_01M_BC'] < 250 and df['IND_FUND'] == 0:
    return '1'
  elif df['VL_ALAV_SV_01M_BC_MED'] < 0.01:
    return '1'
  elif df['VL_ALAV_SV_01M_BC_MED'] < 0.20: 
    return '2'
  else:
    return '3'

def pilar_ala_ind_gde(df):
  if df['VL_ALAVANC_01M_BC'] < 250 and df['IND_FUND'] == 1:
    return '1'
  elif df['VL_ALAVANC_01M_BC'] < 250 and df['IND_FUND'] == 0:
    return '1'
  elif df['VL_ALAV_SV_01M_BC_GDE'] < 0.01:
    return '1'
  elif df['VL_ALAV_SV_01M_BC_GDE'] < 0.20: 
    return '2'
  else:
    return '3'


```

```python
df['PILAR_JORNR_ALA_IND_MIC'] = df.apply(pilar_ala_ind_mic, axis = 1)
df['PILAR_JORNR_ALA_IND_PQN'] = df.apply(pilar_ala_ind_pqn, axis = 1)
df['PILAR_JORNR_ALA_IND_MED'] = df.apply(pilar_ala_ind_med, axis = 1)
df['PILAR_JORNR_ALA_IND_GDE'] = df.apply(pilar_ala_ind_gde, axis = 1)
```
